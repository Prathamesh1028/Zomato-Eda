```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt 
%matplotlib inline
```

```python
df=pd.read_csv("/kaggle/input/zomato-eda/zomato.csv")
df.head()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df=df.drop(['menu_item','listed_in(city)','phone','url','address','reviews_list'],axis=1)
```

```python
df=df.drop_duplicates()
```

```python
df=df.rename(columns={"rate":"ratings","name":"rest name","approx_cost(for two people)":"cost[2 person]","dish_liked":"special dishes"})
```

```python
df.nunique()
```

```python
df.isnull().sum()
```

```python
df['special dishes'].fillna('Pasta',inplace=True)
```

```python
df['ratings'].fillna('-',inplace=True)
df['ratings']=df['ratings'].str.replace('/5|NEW','0',regex=True)
```

```python
T=df.loc[df['ratings']=='-'].index   #  or use isin --boolean indexing 
df=df.drop(T)
```

```python
df['cost[2 person]'] = df["cost[2 person]"].str.replace(',','')
```
